```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Sales target.csv")
```

```python
print(df.columns)
```

```python
df["Month of Order Date"]=pd.to_datetime(df["Month of Order Date"],dayfirst=True)
```

```python
df["Month"]=df["Month of Order Date"].dt.month
```

```python
result=(df.groupby(["Month","Category"],as_index=False)
["Target"].sum())
```

```python
print("\n Monthly Target by Category:\n")
print(result)
result.to_csv("monthly_target_by_category.csv",index=False)
```

```python
for  Category in result["Category"].unique():
    data=result[result["Category"]==Category]
    plt.plot(data["Month"],data["Target"],label=Category)
    plt.xlabel("Month")
    plt.ylabel("Target")
    plt.title("Monthly sales target by category")
    plt.legend()
    plt.show()
```
